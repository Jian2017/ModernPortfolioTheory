```python
# stock and market value
s1 = s[stocknamelist].values[0:60,:]
m1 = m["^GSPC"].values[0:60]
s2 = s[stocknamelist].values[60:120,:]
m2 = m["^GSPC"].values[60:120]
s3 = s[stocknamelist].values[120:180,:]
m3 = m["^GSPC"].values[120:180]
```

```python
print(s3.shape,  m3.shape )
```

```python
# stock and market return
sr1 = (s1[1:60,:] - s1[0:59,:])/s1[0:59,:]
mr1 = (m1[1:60] - m1[0:59])/m1[0:59]
sr2 = (s2[1:60,:] - s2[0:59,:])/s2[0:59,:]
mr2 = (m2[1:60] - m2[0:59])/m2[0:59]
sr3 = (s3[1:60,:] - s3[0:59,:])/s3[0:59,:]
mr3 = (m3[1:60] - m3[0:59])/m3[0:59]
```

```python
# given stock return and market return, compute beta
def get_beta( stock_return , market_return ):
    n = len(market_return)
    market_return_mean = np.mean(market_return)
    stock_return_mean = np.mean(stock_return)
    a = np.sum( ( stock_return -stock_return_mean )*(market_return -market_return_mean ) )
    b = np.sum( (market_return -market_return_mean)*(market_return -market_return_mean ) )
    return a/b
```

```python
get_beta(sr1[:,1],mr1)
```

```python
beta1 = []
beta2 = []
beta3 = []
for i in range(44):
    beta1.append( get_beta(sr1[:,i],mr1) )
    beta2.append( get_beta(sr2[:,i],mr2) )
    beta3.append( get_beta(sr3[:,i],mr3) )
```

```python
dat0 = pd.DataFrame({"stock name":stocknamelist})
dat1 = pd.DataFrame({"beta in period 1": beta1})
dat2 = pd.DataFrame({"beta in period 2": beta2})
dat3 = pd.DataFrame({"beta in period 3": beta3})
dat = dat0.join(dat1).join(dat2).join(dat3)
dat
```

# d. Mean Square Error Decomposition

```python
def error_decomp( As , Ps ):
    
    n = len(As)
    
    A_mean = np.mean(As)
    P_mean = np.mean(Ps)
    
    A_sigma2 = np.mean( (As - A_mean)**2  )
    P_sigma2 = np.mean( (Ps - P_mean)**2  )   
    AP_sigma = np.mean( (As - A_mean)*(Ps - P_mean) )
    
    beta = AP_sigma/P_sigma2
    R2 = AP_sigma/A_sigma2
    
    bias = (A_mean - P_mean)**2
    inefficiency = (1 - beta)**2 * P_sigma2
    random = (1-R2) * A_sigma2
    
    
    #print( np.mean( (As- Ps)**2)  - bias - inefficiency - random )
    print(["bias", "inefficiency", "random"])
    return bias, inefficiency, random
    #A.K.A. bias, variance, random
    
```
