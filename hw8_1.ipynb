```python
def monteCarlo(s0 , mu, sigma , num , pathn):
    out = [ [0]*(num+1) for _ in range(pathn)]
    for i in range(pathn):
        out[i][0] = s0
        temp = s0
        for j in range(1,1+num):
            temp = temp * (1+np.random.normal(mu, sigma))
            out[i][j] = temp 
    return np.array(out)
```

```python
import matplotlib.pyplot as plt
import numpy as np
paths = monteCarlo(50, 0.0038, 0.034, 52,200)
print("number of path = ", paths.shape[0])
plt.plot( np.transpose(paths) )
plt.xlabel("weeks")
plt.ylabel("price of the stock")
plt.title("monte carlo prediction of stock price path")
plt.show()
```
