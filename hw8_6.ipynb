```python
import pandas as pd 
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
a=pd.read_csv("hw8_6.csv")
s = a["AAPL"].values
plt.plot(s,'.-')
plt.xlabel("date in March 2019")
plt.ylabel("stock price of Apple")
plt.show()
```

```python
n = len(s)
print("available days = ",n)
logs = []
for i in range(n-5):
    print(i,i+5)
    logs.append( np.log( s[i+5]/s[i] ) )
nlogs = len(logs)
print("number of log ratio = ", nlogs )
logs_mean_est = np.mean(logs)
logs_sigma_est = np.std(logs) * np.sqrt( nlogs/(nlogs - 1) )
```

```python
sigma_1 = logs_sigma_est / np.sqrt(7)
print("day sigma = " , sigma_1 )
sigma_365 = sigma_1 * np.sqrt(365)
print("year sigma = " , sigma_365 )
```
